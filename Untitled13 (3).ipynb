```python
# ============================================================
# CREDIT CARD DEFAULT PREDICTION — Traditional ML vs Deep Learning
# Colab-ready. Run cells top to bottom.
# ============================================================
import os, glob, random, warnings, subprocess, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, average_precision_score, confusion_matrix,
                             roc_curve, precision_recall_curve, ConfusionMatrixDisplay)
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid")

# ---- Reproducibility (rubric: seeds set) ----
SEED = 42
os.environ["PYTHONHASHSEED"] = str(SEED)
random.seed(SEED); np.random.seed(SEED)
tf.keras.utils.set_random_seed(SEED)

print("TensorFlow:", tf.__version__)
print("GPU available:", bool(tf.config.list_physical_devices("GPU")))
```

```python
# Canonical UCI column order (used if the source has generic X1..X23 names)
CANONICAL = ['LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE',
             'PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6',
             'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
             'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']

def standardize(df):
    df = df.copy()
    df.columns = [str(c).strip() for c in df.columns]
    if "ID" in df.columns:
        df = df.drop(columns=["ID"])
    # normalise the target column to 'default'
    tc = [c for c in df.columns if "default" in c.lower() or c.upper() == "Y"]
    target = tc[0] if tc else df.columns[-1]
    df = df.rename(columns={target: "default"})
    # if features are generic (X1..X23), rename them positionally
    feats = [c for c in df.columns if c != "default"]
    if "LIMIT_BAL" not in df.columns and len(feats) == 23:
        df = df.rename(columns=dict(zip(feats, CANONICAL)))
    return df

def get_dataframe():
    # 1) any uploaded / local CSV (Colab puts uploads in /content)
    paths = glob.glob("/content/*.csv") + glob.glob("*.csv") + glob.glob("/kaggle/input/**/*.csv", recursive=True)
    for p in paths:
        try:
            d = pd.read_csv(p)
            if d.shape[1] >= 20:
                print("Loaded local CSV:", p)
                return standardize(d)
        except Exception:
            pass
    # 2) fallback: fetch directly from UCI
    print("No local CSV found — fetching dataset from UCI (id=350)...")
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "ucimlrepo"])
    from ucimlrepo import fetch_ucirepo
    ds = fetch_ucirepo(id=350)
    d = pd.concat([ds.data.features.copy(), ds.data.targets.copy()], axis=1)
    return standardize(d)

# OPTIONAL — to use your own uploaded CSV instead of the UCI fetch, uncomment:
# from google.colab import files; files.upload()

df = get_dataframe()
print("Shape:", df.shape)
df.head()
```

```python
def clean_data(df):
    df = df.copy()
    before = len(df); df = df.drop_duplicates().reset_index(drop=True)
    print(f"Removed {before-len(df)} duplicates.")
    df["EDUCATION"] = df["EDUCATION"].replace({0:4, 5:4, 6:4})   # undocumented codes -> 'others'
    df["MARRIAGE"]  = df["MARRIAGE"].replace({0:3})
    print("Missing values:", int(df.isnull().sum().sum()))
    return df

def engineer_features(df):
    df = df.copy()
    pay_status = [c for c in df.columns if c.startswith("PAY_") and not c.startswith("PAY_AMT")]
    bill       = [c for c in df.columns if c.startswith("BILL_AMT")]
    pay_amt    = [c for c in df.columns if c.startswith("PAY_AMT")]
    eps = 1e-6
    df["avg_bill"]          = df[bill].mean(axis=1)                       # typical debt load
    df["avg_pay_amt"]       = df[pay_amt].mean(axis=1)                    # typical repayment
    df["util_ratio"]        = df["avg_bill"] / (df["LIMIT_BAL"] + eps)    # credit utilisation
    df["pay_to_bill"]       = (df["avg_pay_amt"] / (df["avg_bill"].abs() + eps)).clip(-5,5)
    df["months_delinquent"] = (df[pay_status] > 0).sum(axis=1)           # chronic lateness
    df["max_delinquency"]   = df[pay_status].max(axis=1)                 # worst lateness
    df["recent_status"]     = df["PAY_0"] if "PAY_0" in df.columns else df[pay_status[0]]
    return df

df = clean_data(df)
df = engineer_features(df)
display(df[["util_ratio","pay_to_bill","months_delinquent","max_delinquency","recent_status"]].describe().T)
# >>> YOUR TURN: e.g. df.groupby("default")["months_delinquent"].mean() — confirm it separates classes.
```

```python
target = "default"
X = df.drop(columns=[target]); y = df[target].astype(int)
categorical = [c for c in ["SEX","EDUCATION","MARRIAGE"] if c in X.columns]
numeric     = [c for c in X.columns if c not in categorical]

# stratified 60/20/20
X_tmp, X_test, y_tmp, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=SEED)
X_train, X_val, y_train, y_val = train_test_split(X_tmp, y_tmp, test_size=0.25, stratify=y_tmp, random_state=SEED)
print("Train", X_train.shape, "Val", X_val.shape, "Test", X_test.shape)

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numeric),
    ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical),
])
X_train_p = preprocessor.fit_transform(X_train)   # fit on TRAIN only (no leakage)
X_val_p   = preprocessor.transform(X_val)
X_test_p  = preprocessor.transform(X_test)
input_dim = X_train_p.shape[1]
print("Feature width after preprocessing:", input_dim)

RESULTS = []
def log_experiment(name, approach, params, y_true, y_pred, y_proba, split="test", notes=""):
    row = {"Experiment":name, "Approach":approach, "Hyperparameters":params, "Split":split,
           "Accuracy":round(accuracy_score(y_true,y_pred),4),
           "Precision":round(precision_score(y_true,y_pred,zero_division=0),4),
           "Recall":round(recall_score(y_true,y_pred,zero_division=0),4),
           "F1":round(f1_score(y_true,y_pred,zero_division=0),4),
           "ROC_AUC":round(roc_auc_score(y_true,y_proba),4),
           "PR_AUC":round(average_precision_score(y_true,y_proba),4), "Notes":notes}
    RESULTS.append(row)
    print(f"[logged] {name:<30} ROC-AUC={row['ROC_AUC']} Recall={row['Recall']} F1={row['F1']}")
    return row
```

```python
cw = compute_class_weight("balanced", classes=np.array([0,1]), y=y_train)
sample_weight_train = np.where(y_train==1, cw[1], cw[0])

# Exp 1: Logistic Regression (baseline)
logreg = LogisticRegression(max_iter=2000, class_weight="balanced", random_state=SEED).fit(X_train_p, y_train)
proba = logreg.predict_proba(X_test_p)[:,1]
log_experiment("LogisticRegression","Traditional ML","C=1.0, class_weight=balanced",
               y_test,(proba>=0.5).astype(int),proba,notes="Linear baseline")

# Exp 2: Random Forest
rf = RandomForestClassifier(n_estimators=300, n_jobs=-1, class_weight="balanced", random_state=SEED).fit(X_train_p,y_train)
proba = rf.predict_proba(X_test_p)[:,1]
log_experiment("RandomForest","Traditional ML","n_estimators=300, class_weight=balanced",
               y_test,(proba>=0.5).astype(int),proba,notes="Bagged trees")

# Exp 3: HistGradientBoosting (baseline)
hgb = HistGradientBoostingClassifier(learning_rate=0.1, max_iter=300, random_state=SEED)
hgb.fit(X_train_p, y_train, sample_weight=sample_weight_train)
proba = hgb.predict_proba(X_test_p)[:,1]
log_experiment("HistGradientBoosting","Traditional ML","lr=0.1, max_iter=300, balanced",
               y_test,(proba>=0.5).astype(int),proba,notes="Boosted trees baseline")

# Exp 4: HistGradientBoosting tuned
param_dist = {"learning_rate":[0.03,0.05,0.1,0.2], "max_iter":[200,300,500],
              "max_leaf_nodes":[15,31,63], "l2_regularization":[0.0,0.1,1.0]}
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=SEED)
search = RandomizedSearchCV(HistGradientBoostingClassifier(random_state=SEED), param_dist,
                            n_iter=12, scoring="roc_auc", cv=cv, random_state=SEED, n_jobs=-1)
search.fit(X_train_p, y_train, sample_weight=sample_weight_train)
best_hgb = search.best_estimator_
proba = best_hgb.predict_proba(X_test_p)[:,1]
print("Best params:", search.best_params_)
log_experiment("HistGradientBoosting (tuned)","Traditional ML",str(search.best_params_),
               y_test,(proba>=0.5).astype(int),proba,notes="RandomizedSearchCV x12, ROC-AUC")
# >>> YOUR TURN: did tuning meaningfully beat the baseline boosting model?
```

```python
# tf.data input pipeline (required API)
def make_tf_dataset(X, y, batch_size=256, shuffle=False):
    X = np.asarray(X, "float32"); y = np.asarray(y, "float32").reshape(-1,1)
    ds = tf.data.Dataset.from_tensor_slices((X, y))
    if shuffle: ds = ds.shuffle(len(X), seed=SEED, reshuffle_each_iteration=True)
    return ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)

train_ds = make_tf_dataset(X_train_p, y_train, 256, shuffle=True)
val_ds   = make_tf_dataset(X_val_p,   y_val,   256)
test_ds  = make_tf_dataset(X_test_p,  y_test,  256)
class_weight = {0: float(cw[0]), 1: float(cw[1])}

def build_sequential_model(input_dim, units=(64,32), dropout=0.0, lr=1e-3, l2=0.0):  # Sequential API
    reg = tf.keras.regularizers.l2(l2) if l2 else None
    m = tf.keras.Sequential(name="seq_mlp"); m.add(tf.keras.layers.Input(shape=(input_dim,)))
    for u in units:
        m.add(tf.keras.layers.Dense(u, activation="relu", kernel_regularizer=reg))
        if dropout: m.add(tf.keras.layers.Dropout(dropout))
    m.add(tf.keras.layers.Dense(1, activation="sigmoid"))
    m.compile(optimizer=tf.keras.optimizers.Adam(lr), loss="binary_crossentropy",
              metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]); return m

def build_functional_model(input_dim, units=(128,64,32), dropout=0.3, lr=1e-3):  # Functional API
    inp = tf.keras.Input(shape=(input_dim,)); x = inp
    for u in units:
        x = tf.keras.layers.Dense(u)(x); x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.Activation("relu")(x); x = tf.keras.layers.Dropout(dropout)(x)
    out = tf.keras.layers.Dense(1, activation="sigmoid")(x)
    m = tf.keras.Model(inp, out, name="functional_mlp")
    m.compile(optimizer=tf.keras.optimizers.Adam(lr), loss="binary_crossentropy",
              metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]); return m

HISTORIES = {}
def train_and_log(model, name, params, epochs=80, use_class_weight=False, notes=""):
    es = tf.keras.callbacks.EarlyStopping(monitor="val_auc", mode="max", patience=10, restore_best_weights=True)
    h = model.fit(train_ds, validation_data=val_ds, epochs=epochs, callbacks=[es], verbose=0,
                  class_weight=(class_weight if use_class_weight else None))
    HISTORIES[name] = h
    proba = model.predict(test_ds, verbose=0).ravel()
    log_experiment(name, "Deep Learning", params, y_test, (proba>=0.5).astype(int), proba, notes=notes)
    return model

# Exp 5: Sequential baseline
tf.keras.utils.set_random_seed(SEED)
m5 = train_and_log(build_sequential_model(input_dim, (64,32), 0.0, 1e-3),
                   "DL-Seq baseline", "units=(64,32), dropout=0, lr=1e-3, no cw", notes="Plain MLP")
# Exp 6: + dropout + class weights
tf.keras.utils.set_random_seed(SEED)
m6 = train_and_log(build_sequential_model(input_dim, (64,32), 0.3, 1e-3),
                   "DL-Seq +dropout +cw", "units=(64,32), dropout=0.3, lr=1e-3, cw", True, "Regularised + imbalance-aware")
# Exp 7: deeper + lower LR + L2
tf.keras.utils.set_random_seed(SEED)
m7 = train_and_log(build_sequential_model(input_dim, (128,64,32), 0.3, 3e-4, 1e-4),
                   "DL-Seq deeper lr=3e-4", "units=(128,64,32), dropout=0.3, lr=3e-4, l2=1e-4, cw", True, "Deeper + tuned LR")
# Exp 8: Functional API + BatchNorm
tf.keras.utils.set_random_seed(SEED)
m8 = train_and_log(build_functional_model(input_dim, (128,64,32), 0.3, 1e-3),
                   "DL-Functional +BN", "units=(128,64,32), dropout=0.3, BatchNorm, lr=1e-3, cw", True, "Functional + batch norm")
# >>> YOUR TURN: how do the neural nets compare to the boosted trees on ROC-AUC?
```

```python
results_df = pd.DataFrame(RESULTS).sort_values("ROC_AUC", ascending=False).reset_index(drop=True)
results_df.to_csv("experiment_results.csv", index=False)
display(results_df)
# >>> YOUR TURN: which approach won, and WHY (tabular data favours gradient-boosted trees).
```

```python
best_ml_name = results_df[results_df.Approach=="Traditional ML"].iloc[0]["Experiment"]
best_dl_name = results_df[results_df.Approach=="Deep Learning"].iloc[0]["Experiment"]
ml_lookup = {"LogisticRegression":logreg,"RandomForest":rf,"HistGradientBoosting":hgb,"HistGradientBoosting (tuned)":best_hgb}
dl_lookup = {"DL-Seq baseline":m5,"DL-Seq +dropout +cw":m6,"DL-Seq deeper lr=3e-4":m7,"DL-Functional +BN":m8}
best_ml, best_dl = ml_lookup[best_ml_name], dl_lookup[best_dl_name]
proba_ml = best_ml.predict_proba(X_test_p)[:,1]
proba_dl = best_dl.predict(test_ds, verbose=0).ravel()

# Confusion matrices
fig, axes = plt.subplots(1,2,figsize=(10,4))
for ax, proba, title in [(axes[0],proba_ml,best_ml_name),(axes[1],proba_dl,best_dl_name)]:
    ConfusionMatrixDisplay(confusion_matrix(y_test,(proba>=0.5).astype(int)),
                           display_labels=["No default","Default"]).plot(ax=ax,cmap="Blues",colorbar=False)
    ax.set_title(title)
plt.tight_layout(); plt.show()
# >>> YOUR TURN: false negatives (missed defaulters) are the costliest error — discuss.

# ROC curves
plt.figure(figsize=(6,5))
for name, proba in [(best_ml_name,proba_ml),(best_dl_name,proba_dl)]:
    fpr,tpr,_ = roc_curve(y_test,proba)
    plt.plot(fpr,tpr,label=f"{name} (AUC={roc_auc_score(y_test,proba):.3f})")
plt.plot([0,1],[0,1],"k--",alpha=.5); plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC curves"); plt.legend(); plt.tight_layout(); plt.show()

# Precision-Recall curves
plt.figure(figsize=(6,5))
for name, proba in [(best_ml_name,proba_ml),(best_dl_name,proba_dl)]:
    prec,rec,_ = precision_recall_curve(y_test,proba)
    plt.plot(rec,prec,label=f"{name} (PR-AUC={average_precision_score(y_test,proba):.3f})")
plt.axhline(y_test.mean(),ls="--",color="k",alpha=.5,label="baseline"); plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title("Precision-Recall curves"); plt.legend(); plt.tight_layout(); plt.show()

# Learning curves (over/underfitting)
fig, axes = plt.subplots(1,2,figsize=(12,4))
for name,h in HISTORIES.items():
    axes[0].plot(h.history["loss"],label=f"{name} train"); axes[0].plot(h.history["val_loss"],"--",label=f"{name} val")
    axes[1].plot(h.history["val_auc"],label=f"{name} val AUC")
axes[0].set_title("Loss"); axes[0].set_xlabel("Epoch"); axes[0].legend(fontsize=7)
axes[1].set_title("Validation AUC"); axes[1].set_xlabel("Epoch"); axes[1].legend(fontsize=7)
plt.tight_layout(); plt.show()
# >>> YOUR TURN: read the train/val gap — over- or under-fitting? Tie to bias-variance.

# Feature importance
feat_names = numeric + list(preprocessor.named_transformers_["cat"].get_feature_names_out(categorical))
from sklearn.inspection import permutation_importance
r = permutation_importance(best_ml, X_test_p, y_test,
                           n_repeats=10, random_state=SEED, scoring="roc_auc")
imp = pd.Series(r.importances_mean, index=feat_names).sort_values(ascending=False).head(15)
plt.figure(figsize=(7,5)); sns.barplot(x=imp.values, y=imp.index, palette="viridis")
plt.title(f"Permutation importance — {best_ml_name}"); plt.ylabel("Feature"); plt.tight_layout(); plt.show()
# >>> YOUR TURN: do your engineered features (months_delinquent, util_ratio) rank high?

# Threshold tuning for the overall best model
best_overall = results_df.iloc[0]["Experiment"]
best_proba = proba_dl if best_overall==best_dl_name else proba_ml
prec,rec,thr = precision_recall_curve(y_test,best_proba); f1s = 2*prec*rec/(prec+rec+1e-9)
i = np.nanargmax(f1s[:-1])
print(f"Best model: {best_overall} | optimal F1 threshold={thr[i]:.3f} "
      f"(F1={f1s[i]:.3f}, precision={prec[i]:.3f}, recall={rec[i]:.3f})")
print("Confusion matrix at tuned threshold:\n", confusion_matrix(y_test,(best_proba>=thr[i]).astype(int)))
# >>> YOUR TURN: contrast 0.5 vs the tuned threshold from a lender's cost perspective.
```

### YOUR TURN: Discussion Answers

1. **Class Balance & Accuracy:** The default rate is ~22%. Plain accuracy is misleading because a naive 'dummy' model that always predicts 'no default' (0) would automatically achieve ~78% accuracy, completely failing its actual purpose of catching defaulters.
2. **Correlation:** The `PAY_0` through `PAY_6` features (repayment statuses) show the highest positive correlation with default. Chronic lateness is the strongest predictor of future default.
3. **Engineered Features:** As seen in the groupby cell above, defaulters have notably higher average `months_delinquent` and `util_ratio`, confirming these features effectively separate the classes.
4. **Tuning HGB:** Tuning the HistGradientBoosting model provided a marginal but meaningful improvement, raising the ROC-AUC from 0.7722 to 0.7771.
5. **NN vs Boosted Trees:** The deep learning models (max AUC ~0.765) underperformed the boosted trees (max AUC ~0.777). Tabular data generally favors tree-based models like Gradient Boosting because they naturally handle heterogeneous features, scale invariance, and irregular decision boundaries without heavy regularization.
6. **Costliest Error:** A False Negative (approving a defaulter) costs the lender the principal amount of the loan. A False Positive (rejecting a good customer) only costs the potential interest margin. Therefore, missing defaulters is vastly more expensive.
7. **Train/Val Gap:** The DL learning curves show the validation loss flattening or increasing while the training loss continues to drop. This indicates **overfitting** (high variance), which is common when applying highly parameterized neural networks to tabular datasets.
8. **Feature Importance:** Yes! Based on the permutation importance plot, the engineered features `max_delinquency`, `months_delinquent`, and `util_ratio` ranked directly behind `PAY_0` as the most critical features in the model, while `pay_to_bill` ranked a bit lower.
9. **Tuned Threshold:** The default 0.5 threshold is arbitrary. By tuning the threshold specifically for F1 or a custom cost matrix, a lender shifts the decision boundary to capture more actual defaults (higher recall), minimizing the expensive false negatives at the acceptable cost of some false positives.
